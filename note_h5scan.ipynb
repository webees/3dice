```python
import os
import pandas as pd
import tensorflow as tf
from IPython.display import display

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
```

```python
dir = os.path.join(os.getcwd(), 's1')

data = []
for file in os.listdir(dir):
    if file.endswith('.h5'):
        model = tf.keras.models.load_model(os.path.join(dir, file))
        num_layers = len(model.layers) - 2
        row = {
            'FILE': file,
            'LAYERS': num_layers,
            'INPUT': model.layers[0].activation.__name__,
        }
        for i in range(1, 8):
            activ = ''
            units = ''
            if len(model.layers) > i:
                if hasattr(model.layers[i], 'activation'):
                    activ = model.layers[i].activation.__name__
                if isinstance(model.layers[i], tf.keras.layers.Flatten):
                    activ = 'FLATTEN'
                if isinstance(model.layers[i], tf.keras.layers.BatchNormalization):
                    activ = 'BATCH'
                units = model.layers[i].output_shape[1]
            row[f'【{i}】ACTIV'] = activ
            row[f'【{i}】UNITS'] = units
        data.append(row)

df = pd.DataFrame(data)

df_sorted = df.sort_values('LAYERS')
display(df_sorted)
```
